## Step 4: Data Preprocessing

```python

# Replace incorrect data in the 'height' column with random values between 150 and 180
data['Height'] = np.random.randint(150, 180, size=len(data))
data
```

```python
# Correcting the 'college' column
# Assuming 'college' column contains categorical data and replacing incorrect values with 'Unknown'
data['College'].fillna("Unknown",inplace=True)
data
```

```python
# Correcting the 'Salary' column
# Calculate the mean of valid salary values
mean_salary = round(data['Salary'].mean())
mean_salary
```

```python
# Fill NaN values in 'Salary' column with the rounded mean salary
data['Salary'].fillna(mean_salary, inplace=True)
data
```

### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python

# Count the number of employees in each team

team_counts = data['Team'].value_counts()
team_counts
```

```python
# Calculate the percentage splitting with respect to the total employees
team_percentage = round((team_counts / len(data)) * 100,2)
team_percentage
```

```python
# Create a new DataFrame with the results
team_table = pd.DataFrame({'Team': team_counts.index, 'Number of Employees': team_counts.values, 'Percentage': team_percentage.values})
team_table
```

- The table provides insights into the distribution of employees across different teams, showcasing the number of employees in each team and the corresponding percentage split with respect to the total employees. Notably, the "New Orleans Pelicans" have the highest number of employees, comprising approximately 4.15% of the total workforce. On the other hand, teams like the "Minnesota Timberwolves" and "Orlando Magic" have  lower percentage, each representing around 3.06% of the total employees. This information allows for a comprehensive understanding of the workforce distribution, aiding in organizational planning and resource management.

```python
# Plotting the bar chart with Seaborn

plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Percentage', data=team_table, palette='viridis')
plt.title('Percentage of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```

<b>Data story:The graph displays the percentage of employees in each team, providing a quick overview of the distribution of employees across different teams in  ABC company.

<b>Data story:Each bar represents a specific age group, and the height of the bar corresponds to the number of employees in that age group. The x-axis shows the age values, and the y-axis represents the count of employees

```python
#Age group analysis
age_groups = pd.cut(data['Age'], bins=[15, 20, 25, 30,35, 40, 45], labels=['15-20', '20-25', '25-30', '30-35', '35-40','40-45'])
age_group_counts = age_groups.value_counts().reset_index()
age_group_counts.columns=['Age group','Number of Employees']
age_group_counts
```

- The table showcases the segmentation of employees into distinct age groups. The 'Age group' column represents various age ranges, such as '20-25,' '25-30,' '30-35,' '35-40,' '15-20,' and '40-45.' The corresponding 'Number of Employees' column reveals the count of individuals within each age group. There are 179 employees aged between 20 and 25, 167 employees aged 25 to 30, and so on. This provides a concise overview of how the workforce is distributed across different age categories at ABC Company.

```python
# Plotting the age group distribution using barplot

plt.figure(figsize=(10, 6))
sns.barplot(x='Age group', y='Number of Employees', data=age_group_counts, palette='colorblind')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

<b>Data story:The resulting graph will have bars representing different age values on the x-axis, and the height of each bar indicates the number of employees in each corresponding age group. The larger the bar, the more employees belong to that age group. The y-axis shows the count of employees, and the x-axis shows the different age values.

- 20-25 Age Group: This age group has the highest count, with 179 employees. It means that the majority of employees fall into the age range of 20 to 25.

- 25-30 Age Group: The second-highest count is in the 25-30 age group, with 167 employees.

- 30-35 Age Group: There are 68 employees in the 30-35 age group.

- 35-40 Age Group: The count decreases further to 23 employees in the 35-40 age group.

- 15-20 Age Group: There are 21 employees in the 15-20 age group.

- 40-45 Age Group: There are no employees in the 40-45 age group (count is 0).

In summary, the 20-25 age group contains the highest number of employees, while the 40-45 age group has no employees. The age groups are ordered by count, from the highest to the lowest.

### 4.Find out under which team and position, spending in terms of salary is high.

```python
# Group the salary by Team and Position, and calculate the sum of salaries for each group
team_position_salary = data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
team_position_salary 

```

```python
# Team and position, spending in terms of salary is high.
max_total_salary_index = team_position_salary['Salary'].idxmax()
max_total_salary_row = team_position_salary.loc[max_total_salary_index]
max_total_salary_row
```

```python
#Display in descending order of salary
team_position_total_desc = team_position_salary.sort_values(by='Salary', ascending=False)
team_position_total_desc.columns = ['Team', 'Position', 'Total Salary Spending']
team_position_total_desc.reset_index(drop=True, inplace=True)
team_position_total_desc
```

- This indicates that the team "Los Angeles Lakers" with the position "SF" has the highest total salary(31866445.0) spending,"Orlando Magic"  with the position "SF" has  comparatively lower total salary(845059.0) spending. This information provides insights into the financial allocation across teams and positions, aiding strategic decision-making related to resource distribution within the organization.

```python
# Create scatterplot plot for total salary spending across different teams and positions

plt.figure(figsize=(14, 8))
sns.scatterplot(x='Team', y='Total Salary Spending', hue='Position',data=team_position_total_desc)
plt.title('Total Salary Spending Across Teams and Positions')
plt.xlabel('Team')
plt.ylabel('Total Salary Spending')
plt.xticks(rotation=90)
plt.show()
```

<b>Data story:
The scatter plot visually represents the relationship between Team and Salary for employees in the dataset

- X-Axis (Team): The x-axis represents the different teams within the organization. Each distinct team is labeled along the x-axis.

- Y-Axis (Salary): The y-axis represents the salaries of individual employees. The position along the y-axis corresponds to the salary of each employee.

- It represent the total salary spending for each position within the Los Angeles Lakers team. For example, the value "31866445.0" in the "Total Salary Spending" column for the Los Angeles Lakers may indicate the total salary spending for all employees in a particular position 

```python
# Create bar plot for total salary spending across different teams and positions

plt.figure(figsize=(14, 8))
sns.barplot(x='Team', y='Total Salary Spending', hue='Position', data=team_position_total_desc)
plt.title('Total Salary Spending Across Teams and Positions')
plt.xlabel('Team')
plt.ylabel('Total Salary Spending')
plt.xticks(rotation=90)
plt.show()
```

#### Data story:The bar plot above illustrates the total salary spending across different teams and positions within the ABC company. This visualization provides a comprehensive view of how financial resources are distributed among various roles and teams, shedding light on the company's investment in its workforce.

   - The x-axis represents each team within the organization.
   - The y-axis represents the cumulative salary expenditure within the team.
   - The bars are further segmented into different hues, each representing a distinct position within the team.
   - The different colors within each team bar represent various positions.


### 5.Find if there is any correlation between age and salary , represent it visually

```python
#Create a DataFrame with 'Age' and 'Salary' columns

age_salary_data = data[['Age', 'Salary']]
age_salary_data 

```

```python
#Calculate the correlation matrix

correlation_matrix = age_salary_data.corr()
correlation_matrix
```

```python
#Plotting the heatmap

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap: Age vs. Salary')
plt.show()
    
    
```

<b>Data story:
- The heatmap visually represents the correlation matrix.

- The correlation matrix provides insight into the linear relationship between 'Age' and 'Salary'.

- The correlation coefficient of 0.21 indicates a positive but weak correlation between age and salary.

- Values closer to 1 indicate a stronger positive correlation.
